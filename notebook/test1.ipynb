```python
import torch
import torch 
import torch.nn as nn 
import torchvision.transforms as T 
from torch.utils.data import Dataset
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torchvision.utils import save_image
from einops import rearrange 
import timm,glob
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import pandas as pd


class MAE(nn.Module):
    def __init__(self, encoder_name="vit_base_patch16_224"):
        super().__init__()
        self.encoder = timm.create_model(encoder_name, pretrained=False, num_classes=0, in_chans=3)
        self.decoder = nn.Sequential(
            nn.Linear(self.encoder.num_features, 512),
            nn.ReLU(),
            nn.Linear(512, 224 * 224 * 3),  # Reconstruct full image
        )

    def forward(self, x):
        B, C, H, W = x.shape
        z = self.encoder(x)  # works now because we pass full images
        out = self.decoder(z)
        return out.view(B, C, H, W)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
model = MAE().to(device)
```
